```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Check available Matplotlib styles (for debugging)
print("Available Matplotlib styles:", plt.style.available)

# Load the dataset
df = pd.read_csv('wine.csv')

# Task 1: Exploratory Data Analysis (EDA)

# 1. Basic exploration
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nSummary Statistics:")
print(df.describe())
print("\nUnique Types:", df['Type'].unique())

# 2. Examine feature distributions
# Use Seaborn's theme for styling
sns.set_theme(style='whitegrid')

# Histograms
df.drop('Type', axis=1).hist(figsize=(14, 12), bins=20)
plt.suptitle('Feature Distributions')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust for title
plt.show()

# Box plots
plt.figure(figsize=(14, 6))
sns.boxplot(data=df.drop('Type', axis=1))
plt.title('Box Plots of Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Investigate correlations
corr = df.drop('Type', axis=1).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

# Task 2: Dimensionality Reduction with PCA

# 1. Standardize and apply PCA
X = df.drop('Type', axis=1)
y = df['Type']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# PCA loadings
loadings = pd.DataFrame(pca.components_.T, index=X.columns, columns=[f'PC{i+1}' for i in range(pca.n_components_)])
print("\nPCA Loadings:")
print(loadings)

# 2. Determine optimal components
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

# Scree plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()

# Cumulative variance plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.axhline(y=0.8, color='r', linestyle='--', label='80% Variance')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print("\nExplained Variance Ratio:", explained_variance)
print("Cumulative Variance:", cumulative_variance)

# 3. Transform with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Type'] = y
print("\nPCA-Transformed Data (first 5 rows):")
print(pca_df.head())

# Task 3: Clustering with Original Data

# 1. K-means clustering
kmeans_orig = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init for stability
clusters_orig = kmeans_orig.fit_predict(X_scaled)
df['Cluster_Orig'] = clusters_orig

# 2. Visualize (using first two features)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df['Cluster_Orig'], palette='Set1', style=y, size=y, sizes=(50, 150))
plt.title('K-means Clustering on Original Data (Alcohol vs. Malic)')
plt.xlabel('Alcohol (Standardized)')
plt.ylabel('Malic (Standardized)')
plt.legend(title='Cluster/Type', loc='best')
plt.tight_layout()
plt.show()

# 3. Evaluate
silhouette_orig = silhouette_score(X_scaled, clusters_orig)
davies_bouldin_orig = davies_bouldin_score(X_scaled, clusters_orig)
print(f"Original Data - Silhouette Score: {silhouette_orig:.3f}")
print(f"Original Data - Davies-Bouldin Score: {davies_bouldin_orig:.3f}")

# Task 4: Clustering with PCA Data

# 1. K-means clustering
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10) # Added n_init for stability
clusters_pca = kmeans_pca.fit_predict(X_pca)
pca_df['Cluster_PCA'] = clusters_pca

# 2. Visualize
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster_PCA', palette='Set1', style='Type', size='Type', sizes=(50, 150), data=pca_df)
plt.title('K-means Clustering on PCA Data (PC1 vs. PC2)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster/Type', loc='best')
plt.tight_layout()
plt.show()

# 3. Evaluate
silhouette_pca = silhouette_score(X_pca, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(X_pca, clusters_pca)
print(f"PCA Data - Silhouette Score: {silhouette_pca:.3f}")
print(f"PCA Data - Davies-Bouldin Score: {davies_bouldin_pca:.3f}")

# Task 5: Comparison and Analysis
print("\nTask 5: Comparison and Analysis")
print("Original Data Metrics:")
print(f"  Silhouette Score: {silhouette_orig:.3f}")
print(f"  Davies-Bouldin Score: {davies_bouldin_orig:.3f}")
print("PCA Data Metrics:")
print(f"  Silhouette Score: {silhouette_pca:.3f}")
print(f"  Davies-Bouldin Score: {davies_bouldin_pca:.3f}")

# Task 6: Conclusion and Insights
print("\nTask 6: Conclusion and Insights")
print("Key Findings:")
print("- EDA revealed varied feature distributions and correlations (e.g., Flavanoids–Phenols: 0.86).")
print("- PCA with 2 components captured ~55% variance.")
print("- Clustering on PCA data outperformed original data (Silhouette: 0.434 vs. 0.361, Davies-Bouldin: 1.119 vs. 1.394).")
print("Practical Implications:")
print("- PCA reduces dimensionality, improving clustering efficiency.")
print("- Clustering identifies natural groupings, useful for segmentation.")
print("Recommendations:")
print("- Use PCA for high-dimensional data; use original data for interpretability.")
```
